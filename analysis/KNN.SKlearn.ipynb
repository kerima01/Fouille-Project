```python
dfw = pd.read_csv('tb_final.csv')
dfw
```

```python
clf =  neighbors.KNeighborsClassifier( p=1 )
```

```python
# SÃ©parer les variables quantitatives et qualitatives
quantitative_cols = dfw.select_dtypes(include=[np.number]).columns
qualitative_cols = dfw.select_dtypes(exclude=[np.number]).columns
qualitative_cols = qualitative_cols[qualitative_cols != 'HeartDisease'] # ne pas inclure la variable cible

# Afficher les colonnes pour vÃ©rifier
print("Quantitative columns:", quantitative_cols)
print("Qualitative columns:", qualitative_cols)

```

```python
# Encodage des variables qualitatives
dfw_encoded = pd.get_dummies(dfw[qualitative_cols], drop_first=True).astype(int)

# Concatenate les variables quantitatives avec les variables qualitatives encodÃ©es, et la variable cible
dfw_final = pd.concat([dfw[quantitative_cols], dfw_encoded, dfw['HeartDisease']], axis=1)
```

```python
dfw_final
```

```python
# Variables explicatives (X) : toutes les colonnes sauf la cible
x = dfw_final.drop("HeartDisease", axis=1)
x
```

```python
# Variable cible (y) : 'HeartDisease'
y = dfw_final["HeartDisease"].astype('category')  # Transformation en catÃ©gorie
y
```

```python
# SÃ©paration des donnÃ©es en train (2/3) et test (1/3)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
# Afficher les dimensions de X_train, X_test
print("DonnÃ©es d'entraÃ®nement (X) :", x_train.shape)
print("DonnÃ©es de test (X) :", x_test.shape)
```

```python
# Afficher les dimensions de y_train, y_test
print("DonnÃ©es d'entraÃ®nement (y) :", y_train.shape)
print("DonnÃ©es de test (y) :", y_test.shape)
```

```python
#Preparation du model
clf.fit(x_train, y_train)
clf
```

```python
#Evaluation des performances
clf.predict(x_test)
```

```python
# PrÃ©dictions sur les donnÃ©es de test
y_pred = clf.predict(x_test)

# CrÃ©er un DataFrame avec les variables explicatives, les classes rÃ©elles et les classes prÃ©dites
df_results = x_test.copy()  # On commence par copier x_test
df_results['real_class'] = y_test  # Ajouter la colonne des classes rÃ©elles
df_results['predicted_class'] = y_pred  # Ajouter la colonne des classes prÃ©dites

# Afficher les premiÃ¨res lignes pour vÃ©rifier
print(df_results.head())
```

```python
# Calcul du taux de prÃ©cision
accuracy_rate = sum(df_results['real_class'] == df_results['predicted_class']) / len(df_results)
accuracy_rate
```

```python
#taux d'erreur
error_rate = sum(df_results['real_class'] != df_results['predicted_class']) / len(df_results)
error_rate
```

```python
# Matrice de confusion
cm = pd.crosstab(df_results['real_class'], df_results['predicted_class'], rownames=['Classe rÃ©elle'], colnames=['Classe prÃ©dite'])
print("Matrice de confusion :")
print(cm)
```

```python
def confusion_matrix(known, predicted):
    cm = pd.crosstab(known,predicted)
    A = len(known) # for All

    P  = cm.apply(sum, axis=1) # known class is row sums â†’ P
    N = A - P

    PP = cm.apply(sum, axis=0) # predicted class is col sums â†’ PP
    PN = A - PP

    TP = np.diag(cm)
    FP = PP - TP

    FN = P - TP
    TN = A - TP - FN - FP

    sensibility = TP/P
    specificity = TN/N
    
    precision = TP/PP
    recall = TP/P
    f1_score=2*precision*recall/(precision+recall), 
    mcc= (TP*TN-FP*FN) / np.sqrt( (TP+FP)*(TP+FN)*(TN+FP)*(TN+FN)  )
    
    return cm.assign(p=P, 
                     n=N, 
                     pp=PP, 
                     pn=PN, 
                     tp=TP,
                     fp=FP, 
                     tn=TN, 
                     fn=FN, 
                     sensiibility=sensibility, 
                     specificity=specificity, 
                     precision=precision, 
                     recall=recall, 
                     f1_score=np.array(f1_score).T, 
                     mcc=mcc)  
```

```python
np.round( confusion_matrix(df_results['real_class'], df_results['predicted_class']), 2)
```

```python
import sklearn.metrics
print( sklearn.metrics.classification_report(df_results['real_class'], df_results['predicted_class']) )
```

```python
# 1. Boucle pour validation croisÃ©e Ã  3 plis sans scoring spÃ©cifique
print("Validation croisÃ©e avec cv=3 :\n")
for k in range(1, 16):
    clf = neighbors.KNeighborsClassifier(k)
    scores = cross_val_score(clf, x, y, cv=3)
    print(f"RÃ©sultats pour k = {k}:")
    print(f"Scores : {scores}")
    print(f"Score moyen : {scores.mean()}")
    print('-' * 40)
```

```python
# 2. Boucle pour validation croisÃ©e Ã  10 plis avec scoring 'f1_macro'
print("\nValidation croisÃ©e avec cv=10 et scoring='f1_macro':\n")
for k in range(1, 16):
    clf = neighbors.KNeighborsClassifier(k)
    scores = cross_val_score(clf, x, y, cv=10, scoring='f1_macro')
    print(f"RÃ©sultats pour k = {k}:")
    print(f"Scores F1 macro : {scores}")
    print(f"Score moyen F1 macro : {scores.mean()}")
    print('-' * 40)
```
